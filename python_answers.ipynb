# 2.PROPORTIONAL SAMPLING

```python
import random
import bisect
A =[2,6,5,100,20,1]

def proportional_sampling(A,n=100):
    #STEP1
    #compute sum of elements(S)
    S=sum(list(A))
    
    #STEP2
    #Normalizing the values using the sum(A_dash)
    A_dash= [i/S for i in A]
    
    #STEP3
    #compute cummulative normalized values(A_bar)
    A_bar = []
    cumsum = 0
    for ele in A_dash:
        cumsum += ele
        A_bar.append(cumsum)
    
    #STEP4
    #picking an element with the probability propotional to its magnitude 
    num=[]
    
    for i in range(n):
        r = random.uniform(0,1)   
        idx=bisect.bisect(A_bar,r*A_bar[-1])  #reference:stackoverflow
        num.append(A[idx])
        #count_num=num.count(100)  #to check no.of elements present in list
    #print(count_num)
    return num

proportional_sampling(A)
```

# 5.CLOSEST POINTS TO P

```python
import math
def closest_points_to_p(S,P):
    norm_x, dot_xy, norm_y = 0, 0, 0
    cosine=[]
    for a in S:
        for i in range(len(a)):
            x=a[i]
            y=P[i]
            norm_x += x*x
            norm_y += y*y
            dot_xy += x*y
        c=math.acos(dot_xy/math.sqrt(norm_x*norm_y)) #we get in radians
        cosine.append(c)
        cosine_dist=[math.degrees(i) for i in cosine] #converted to degrees
    d=dict(zip(S,cosine_dist))  #dictionary ={datapts,cosine_distance value}
    for key in sorted(d, key=lambda k: d[k])[:5]: #sort dictionary based on cosine dist values and print top5 datapts
        print(key)
    return

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)
closest_points_to_p(S,P)
```

# 6.LINE SEPERATES RED AND BLUE POINTS

```python
import re
import math
def i_am_the_one(Red,Blue,line):
    count1=0
    count2=0
    z= re.findall(r'[\d\.\-\+]+', line)
    for i in range(len(z)):
        z[i]=float(z[i]) #extracted coefficients from equation of line 
        #print(z[i])     #converted list of strings to numbers using re library
    
    
    for i in range(len(Red)):
        dist1=(z[0]*Red[i][0]+z[1]*Red[i][1]+z[2])
        #print(dist1)
        if (dist1>=0): #if dist1 above the line add1
            count1+=1
        elif (dist1==0):#if dist1=0 move to next line
            continue
        else:
            count1-=1 #if dist1 below the line subtract1
     
    for i in range(len(Blue)):
        dist2=(z[0]*Blue[i][0]+z[1]*Blue[i][1]+z[2])
        #print(dist2)
        if (dist2>=0): #if dist2 above the line add1
            count2+=1
        elif (dist2==0): #if dist2=0 move to next line
            continue
        else:
            count2-=1 #if dist2 below the line subtract1
   
    
    #to check signs whether red and blue lie on which side of the line
    n = count1*count2
    if(n<0):
        #print("opp sign") #red and blue point lie on opp side of the line
        if ((abs(count1)==len(Red)) and (abs(count2)==len(Blue))):
            print("yes") 
        else:
            print("no")
        
    else:
        #print("same sign")
        print("no") #red and blue point lie on same side of line

Red= [(1,1),(2,1),(4,2),(2,4), (-1,4)]
Blue= [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]
Lines=["1x+1y+0","1x-1y+0","1x+0y-3","0x+1y-0.5"]

for i in Lines:
    yes_or_no = i_am_the_one(Red, Blue, i)
```

# 8.COMPUTE CONDITIONAL PROBABILITIES

```python
def compute_conditional_probabilites(A):
    
    #split list of lists into two seperate lists
    list1=[i[1] for i in A]
    list2=[i[0] for i in A]
    
    set_list1=list(set(list1))
    set_list2=list(set(list2))
    
    set_list1.sort()
    set_list2.sort()
    
    for F in set_list2:
        for S in set_list1:
            #print(F,S)
    
            num=0
            den=0
            for i in range(len(A)):
                if(A[i][1]==S):
                    den=den+1
                    if(A[i][0]==F):
                        num=num+1
            print('P(F={}|S=={})='.format(F,S),num/den)
    

A = [['F1','S1'],['F2','S2'],['F3','S3'],['F1','S2'],['F2','S3'],['F3','S2'],['F2','S1'],['F4','S1'],['F4','S3'],['F5','S1']]
compute_conditional_probabilites(A)   
```

# 10.ERROR FUNCTION

```python
#Log Loss is a measure of how well a classifier is working
import math
def compute_log_loss(A):
    
    sum= 0
    for y,y_score in A:
        sum += y*math.log10(y_score)+(1-y)*math.log10(1-y_score)  
    log_loss=(-1/len(A))*(sum)  #log loss formula
    
    return log_loss

A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]] #lists of lists

#actual value=y==>[1,0,0,0,0,1,1,1] #binary labels either 0 or 1
#predicted value=y_score==>[0.4,0.5,0.9,0.3,0.6,0.1,0.9,0.8]  #predicted probabilities floats between 0-1

loss = compute_log_loss(A)
print(loss)
```
